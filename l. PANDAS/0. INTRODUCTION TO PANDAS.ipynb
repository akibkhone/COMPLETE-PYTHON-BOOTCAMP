# <CENTER> INTRODUCTION TO PANDAS

1. Pandas is an Open Source Python Library (or) Module Providing hign performance in DATA ANALYSIS and DATA ANALYTICS.
2. The word Pandas derived from a word **"PANel DAta"**.
3. The pandas organising the data Multi-Dimensional Array ( 1-D , 2-D ).
4. The pandas concept developed by **WES MCKINNEY** in the year 2008.
5. The standard python distribution does not contain pandas module. So that pandas module must be installed explicitly by using pip tool.
6. Installation of pandas :

                           pip install pandas
                
7. The Pandas Modules developed in C ( for faster execution as it has best latency ) and PYTHON ( for confortablity of the programming language).

### KEY FEATURES OF PANDAS :

1. Pandas provide Fast and Efficient Data Structures ( Series and DataFrame ) with Default and Customized Indexing.
2. Labelled Slicing and Indexing Operations can be performed.
3. Columns from Data structures ( Series and DataFrame ) can be UPDATED , DELETED and INSERTED.
