##  FEATURES OF PYTHON

- Python programming have the following features :

                            1. Simple
                            2. Freeware and Open Source
                            3. Dynamically Typed Programming
                            4. Platform Independent and portable
                            5. Interpreted
                            6. High Level
                            7. Both Procedural (Functional) and Ibject Oriented Programming language
                            8. Robust (Strong)
                            9. Extensible
                            10. Embedded.
                            11. Supports Extensive Third Party APIs ( Numpy , Pandas , Matplotlib , Scipy , Scikit,....)
                                    

### 1. Simple

- Python is one of the simple programming bcz of 3 important tech factors -

1. Python programming provides rich sets of APIs. So that Python programmmer can re-use the pre-defined code without writing our code.
        - Defination of API ( Application Programming Interface ) -
                1. An Api is a collection modules.
                2. A module is a collection variables, functions and classes.
                
2. Python programming provides in-built garbage collection facility to collect un-used memory space to it improves the performance of python based applications.
    - Defination of Garbage Collection -
        A garbage collection is one of the in-built python background program running behind of every python program                   whose purpose is to collect un-used memory space to it improves the performance of python based applications.
           
           
3. Python programming provides user-friendly syntaxes. So that python programmer can write error-free programs in a limited span of time.

###  3. Dynamically Typed Programming

- We have 2 types of programming language -
        1. Static Typed Programming
        2. Dynamically Typed Programming
        
1. Static typed programming language , it is mandatory to specify the data type for allocating memory space otherwise compile time error.
        example : C , CPP , JAVA , .NET ....etc
        
2. Dynamically typed programming , it is not necessary to specify that data type and data type of the variable is decided by python environment and data type is assigned.
        example : Python
        
- In python programming all the values are stored in the form of objects and behind the objects we have classes.

        Example :  
                    a = 10                  #here 'a' is an object <class, 'int'>
                    b = 10.9                #here 'b' is an object <class, 'float'>
                    print(a, type(a))        #10 <class 'int'>    
                    print(b, type(b))        #10.9 <class 'float'>

###  4. Platform Independent and portable

- In python rogramming all the values are treated as "objects" and objects can stored unlimited number of values and they can run and same on all the OSes (operation systems), and hance python language is considered as "Platform Independent".

EXAMPLES :

1. Platform dependent language : C , CPP... etc, bcz data type of C ,CPP are containing specific memory space can change from OS to another OS.

2. platform Independent Language : Java data types takes same memory space on all OSes but can store only small values and fixed number of values.

PORTABLE : Portable applications / projects are those which are running on every OS with considering their vendors.

            EXAMPLE : Python and JAVA projects are PORTABLE
                      C , CPP projects comes under NON-PORTABLE
                      



###  5. Interpreted

- In python programming, when we RUN the python program, internally python compiler reads line by line of Source Code and converted into byte code (Intermediate Code Python Language) and PVM execute the python program by reading line of byte code and converted into machine understandable code (Binary format or executable code ) and it read by OS and processor and gives result.

- Hence during compilation and execution and line by line operation has performed and python is one of the interpreted programming language.
