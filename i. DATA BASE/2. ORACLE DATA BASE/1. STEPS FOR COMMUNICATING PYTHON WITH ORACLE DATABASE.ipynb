###  STEPS FOR COMMUNICATING PYTHON WITH ORACLE DATABASE

The following steps are used for establisshing the comminication between python and oracle database -


1. import cx_Oracle
2. Python program must get the connection from Oracle Database.
3. Create an Object of Cursor in python program.
4. Design a query, Place it in the object of Cursor, Execute the query in DB SW and gets result from database s/w and Hand          over to the python program .
5. Python program the result.
6. Close the connection.

### 3. Create an Object of Cursor in python program :

The purpose of creating the cursor object is that "To carry the query from python program to database software and bring the result from database software to the python program . Hence an object of cursor acts as driver between python program and database software.

To create an object of cursor, we use a pre-defined function called cursor() which is present in connection class of cx_Oracle module.

SYNTAX :

            varname = connectionobj.cursor()
            
            Here,
                    varname is an object of <class,cursor>
