## <CENTER > 1.  CLASS

- The principle "class" is one of the distinct principle of OOPs.
- **The purpose of "classes" concept is that " To develop Programmer / User defined data types and to develop any real time applications"**. 
- **The main purposed of developing programmer defined data types is that to customized the Operations / Requirements of the problem statements**.
- To develop programmer defined data type and to develop any real time application by using classes concept, we must use a keyword "**class**".


- *DEFINITION OF class* :

 <center>   <b> " A class is a collection of Data Members ( Instance Data Members (specific) and Class Level Data Members (common) )
   
    <center> and Methods ( Functions ) "</b>

- Whatever the data members and methods we are taking inside of class, they are reflecting as a part of object.
- When we define a class, we never get any memory space for Data Members and methods but whise memory space is create when we create an object.
- Hence, Data always stored in object and to create an object, their must exist an class definition, otherwise object creation is not possible.

                                                   Syntax for Defining a Class
                                             ==========================================
     
                                class  <class-name>:

                                        We can Specify Instance Data Members

                                    def  class-level-method(cls,list of formal params if any):
                                            ---------------------------------
                                            We can Specify Class Level Members

###  Types of Data Members in  a Class of Python


- In a class of python, we can define two types of Data Memebrs. They are
		a) Instance Data Members
		b) Class Level Data Members.
        
------------------------------------------------------------------------------

##### a) Instance Data Members

------------------------------------------------------------------------------

- Instance Data Members are used for storing specific Values.
- Instance Data Members memory space cerated every time when we create an 
    object.
- Instance data members must always accesed within Instance Methods only.
- Instance Data members must accessed w.r.t object name (or) self.

			objectname.Instance Data members name
					(or)
			self.Instance Data members name
            
--------------------------------------------------------------------------------------------------------

##### b) Class Level Data Members.

--------------------------------------------------------------------------------------------------------

- Class Level Data Members are used for storing Common Values.
- Class Level Data Members memory space created One time irrespective of number 
     of objects are created.
- Class Level Data Members can be accessed in all types of methods of a class.
- Class Level Data Members can be accessed w.r.t Class name (or) object name 
    (or) self  (or) cls  depends on availability.
    
    
			classname.Class Level Data member name
					(or)
			cls.Class Level Data member name
					(or)
			objectname.Class Level Data member name
					(or)
			self.Class Level Data Level  member name

##  TYPES OF METHODS

- In a class of python, we can define 3 types of Methods. they are -

                1. Instance Method
                2. Class Level Method 
                3. Static Method

###  1. Instance Method

- Instance method are used for performing specific operations on the objects. hence instance methods are also called object level methods.
- Inside of instance methods we can specify instance data members.
- Programatically, Instance Methods always first parameter as "self" , which is always points to current class object.

*SYNTAX* : Instance methods must be called or accessed w.r.t object name.

                objname.Instance Method Name

*SYNTAX for defining instance method in a class is * :

                        def methodname(self, list of formal parameters if any):
                        
                                    --------------------------
                                    block of statements - operations on objects
                                    we can specify instance data members
                                    ---------------------------


```python
class student:
    
    crs = "PYTHON"    #Class Level Data Member
    
    def readstudentdata(self):
        print(f"ID : {id(self)}")
        self.stno = int(input("Enter Student Number :"))
        self.name = input("Enter Student Name :")
        self.marks = float(input("Enter Student Marks :"))
        print("-"*30)
        
    
    def displaystudentdata(self): 
        print("-"*30)
        print(f"ID : {id(self)}")
        print(f"Student Number : {self.stno}")
        print(f"Student Name : {self.name}")
        print(f"Student Marks : {self.marks}")
        print(f"Student Course : {student.crs}")
        
  
#MAIN PROGRAM

s1 = student()
s2 = student()    

print("Enter First Student Information : ")
print("-"*30)
s1.readstudentdata()

print("Enter Second Student Information : ")
print("-"*30)
s2.readstudentdata() 

print("First Student Information :")
s1.displaystudentdata()

print("Second Student Information :")
s2.displaystudentdata()
```

###   2. Class Level Method

- class level methods are used for performing common operations for all objects.
- Inside of clss level methods we can specify class level data members.
- programatically, class level methods definiation must starts with a decorator called **@classmethod** and always first parameter as **cls** , which is always points to corresponding class.

*SYNTAX for defining class level method in a class is* :

                        def methodname(self, list of formal parameters if any):
                        
                                    --------------------------
                                    block of statements - operations on class level operations
                                    we can specify class level data members
                                    ---------------------------
                                    

*SYNTAX* : Class level methods must be called or accessed w.r.t class name (or) self (or) object name.

                                    Classname.Class Level Method 
                                                (OR)
                                    objname.Class Level Method 
                                                (OR)
                                    self.Class Level Method 
                                                (OR)
                                    cls.Class Level Method

##### PROGRAM FOR ACCEPTING 2 VALUES AND ACCEPT ANY ARITHMATIC OPERATOR BY USING OOPS

```python
class Arithmetic: 
    
    def readvalues(self):
        self.a=float(input("Enter 1st Value : "))
        self.b=float(input("Enter 2nd Value : "))
        self.op=input("Enter Any Arithmetic Operator : ")
 

class Calculator:   #### USING STATIC METHOD HERE

    @staticmethod

    def calc(Akib):
        print("-"*30)
        if (Akib.op=="+"):
            print(f"Sum of {Akib.a} and {Akib.b} = {Akib.a + Akib.b}")
        elif (Akib.op=="*"):
            print(f"Multipli of {Akib.a} and {Akib.b} = {Akib.a * Akib.b}")
        elif (Akib.op=="-"):
            print(f"Subtraction of {Akib.a} and {Akib.b} = {Akib.a - Akib.b}")
        elif (Akib.op=="**"):
            print(f"Square of {Akib.a} and {Akib.b} = {Akib.a ** Akib.b}")
        elif (Akib.op=="%"):
            print(f"Modulus of {Akib.a} and {Akib.b} = {Akib.a % Akib.b}")
        elif (Akib.op=="/"):
            print(f"Division of {Akib.a} and {Akib.b} = {Akib.a / Akib.b}")
        else:
            print(f"{Akib.op} is NOT an Arithmetic Operator")
        print("-"*30)
            
#         print("type of Akib = ", type(Akib))
#         print("Content of Akib",Akib.__dict__)print 


# MAIN PROGRAM

ao = Arithmetic()
print("Content of ao before reading = ",ao.__dict__)
ao.readvalues()
print("Content of ao after reading = ",ao.__dict__)
Calculator.calc(ao)
```

WAP which will accept 2 integer values and compute all arithmetic operator with classes and object

```python
class aop: 
    
    def readvalues(self):
        print("id of readvalues = " , id(self))
        self.a=float(input("Enter Value Of a : "))
        self.b=float(input("Enter Value Of b : "))
        
    def aop_operations(self):
        
        x1 = self.a + self.b             # here, x1,x2,x3................. x7 are local variables
        x2 = self.a - self.b
        x3 = self.a * self.b
        x4 = self.a / self.b
        x5 = self.a // self.b
        x6 = self.a % self.b
        x7 = self.a ** self.b
        
        return x1,x2,x3,x4,x5,x6,x7
    
    def displayvalues(self,result):
        print("-"*30)
        print("ARITHMETIC OPERATIONS")
        print("-"*30)
        print(f" SUM OF {self.a} and {self.b} = {result[0]}")
        print(f" SUB OF {self.a} and {self.b} = {result[1]}")
        print(f" MULTIPLICATION OF {self.a} and {self.b} = {result[2]}")
        print(f" DIVISION OF {self.a} and {self.b} = {result[3]}")
        print(f" FLOOR DIVISION OF {self.a} and {self.b} = {result[4]}")
        print(f" MODULUS OF {self.a} and {self.b} = {result[5]}")
        print(f" EXPO OF {self.a} and {self.b} = {result[6]}")
        print("-"*30)
        

# MAIN PROGRAM

ao = aop()
print("id of a in main program = " , id(ao))

#calling function

ao.readvalues()
res = ao.aop_operations()                           ######## res is a tuple which is passing to display function 
ao.displayvalues(res)                               ####### ao for self , res for result 
print(res, "TYPE OF res ",type(res))
```

WAP which will calculate roots of quadratic equation by using classes and object.

```python
class quad:
    
    def readvalues(self):
        self.a=float(input("Enter Value Of a : "))
        self.b=float(input("Enter Value Of b : "))
        self.c=float(input("Enter Value Of c : "))
      
    def calroots(self):  
        self.r1=(- self.b + ( (self.b**2 - 4 * self.a * self.c)**0.5 )) / ( 2 * self.a)
        self.r2=(- self.b - ( (self.b**2 - 4 * self.a * self.c)**0.5 )) / ( 2 * self.a)
    
    def displayresult(self):
        print(f"Value Of a = {self.a}")
        print(f"Value Of b = {self.b}")
        print(f"Value Of c = {self.c}")
        print(f"ROOT 1 = {self.r1}")
        print(f"ROOT 2 = {self.r2}")
        
#MAIN PROGRAM

qo=quad()
qo.readvalues()
qo.calroots()
qo.displayresult()
```

WAP which will calculate factorial of a given number using  class and object.

```python
class fact:
    def read(self):
        self.a = int(input("Enter Any Number To to Calculate its Factorial : "))
        
    def ofac(self):
        fac=1
        for i in range(1,self.a+1):
            fac = fac*i
        print(f"The Factorial of {self.a} is {fac}")
        
#mqin program

f = fact()
f.read()
f.ofac()
    
```

WAP which will accept an integer value and decide whether it is PRIME or NOT using class and object.

```python
class prime:
    
    def read(self):
        self.a = int(input("Enter Any Number To Check Whether it is PRIME or NOT : "))
        
    def oprime(self):   
        
        for i in range(2,self.a):
            
            if (self.a%i) == 0:  
                print(f"{self.a} is not a Prime Number")    
                break
        else:
            print(f"{self.a} is a Prime Number")
       
                
p = prime()
p.read()
p.oprime()
```
