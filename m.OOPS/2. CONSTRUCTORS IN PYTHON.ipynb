##    1. Purpose of  Constructor


- The purpose of constructor is that  " *to initialize the Objects* ".
- Initializing the object is nothing but placing our values without keeping the object empty.

##  2. Definition of  Constructor   

- **A constructor is a special method which is automatically / implicitly called by PVM during object creation and whose role is that to place our own values without keeping object empty ( ie. Initilizing the object ).**

###  PROPERTIES / CHARACTERISTICS / RULES OF CONSTRUCTOR 

- A constructor is a special method which is  automatically / implicitly called by PVM during object creation and it initialize the objects.
- The name of the constructor is **def _ _ init _ _ ( self, formal parameters if any ) : pass**
- Constructor of python can be inherited.
- Constructor of python can be overridden / re-defined / re-written.
- Constructor of python cannot be overloaded.

## 3. Types of  Constructor

- In python programming, we have 2 types of constructor. They are -

                1. Default / Parameterless Constructor
                
                2. Parameterized Constructor


================================================

###  1. Default / Parameterless Constructor

================================================

- A constructor is said to be default constructor if and only if which never takes values / parameter.
- The purpose of default constructor is that to initialize multiple objects of same class with same values.

*SYNTAX -*

                    def __init__(self):
                    
                            -------------------------------------
                            BLOCK OF STATEMENTS - INITIALIZATION
                            -------------------------------------


```python
# default constructor 

class test:
    
    def __init__(self):
        self.a = 10
        self.b = 20
        print(f" Value Of a = {self.a}")
        print(f" Value Of b = {self.b}") 
        print("-------------------")
        
#main program

t1 = test()
t2 = test()
t3 = test()
```

================================================

### 2. Parameterized Constructor

================================================



- A constructor is said to be parametrized constructor if and only if it always takes values / parameter.
- The purpose of default constructor is that to initialize multiple objects of same class with different values.

*SYNTAX -*

                    def __init__( self , list of formal parameters ):
                    
                            -------------------------------------
                            BLOCK OF STATEMENTS - INITIALIZATION
                            -------------------------------------


#### NOTE :

In a single class of python, we can define only one constructor even we create multiple objects with both default and parametrized constructor and it will full-fill the need of both default and parametrized constructor.

```python
# parametrized constructor 

class test:
    
    def __init__(self,a,b):
        self.a = a                         #........... here a,b are called formal parameters
        self.b = b                          # .......... here self.a , self.b are called instance data member
        print(f" Value Of a = {self.a}")
        print(f" Value Of b = {self.b}") 
        print("-------------------")
        
#main program

t1 = test(10,20)
t2 = test(100,200)
t3 = test(1000,2000)
```

```python
# WITHOUT CONSTRUCTOR

class employee: 
    
    def reademp(self):                                          #self is pointing to "eo" object
        self.eno=int(input("Enter Employee Number : "))
        self.ename=input("Enter Employee Name : ")
        self.esal=float(input("Enter Employee Salary : "))


#main program

eo = employee()                            # object is "eo" which always contains INSTANCE DATA MEMBER 

print(f"Content of eo : {eo.__dict__}")                #empty before reading

eo.reademp()

print(f"Content of eo : {eo.__dict__}")
```

```python
# WITH CONSTRUCTOR ( DEFAULT )

class employee: 
    
    def __init__(self):           #constructor syntax                     #self is pointing to "eo" object
        
        self.eno=int(input("Enter Employee Number : "))
        self.ename=input("Enter Employee Name : ")
        self.esal=float(input("Enter Employee Salary : "))


#main program

e1 = employee()                             #here employee() is act as function with constructor call

print(f"Content of e1 : {e1.__dict__}")              # "e1" is object, which always contains INSTANCE DATA MEMBER 


e2 = employee()                             

print(f"Content of e2 : {e2.__dict__}")             
```

```python
# WITH CONSTRUCTOR ( PARAMETRIZED )

class employee: 
    
    def __init__(self,eno,ename,esal):           #constructor syntax     #self is pointing to "eo" object
        
        self.eno=eno
        self.ename=ename
        self.esal=esal


#main program

e1 = employee(1,"Akib",989999)                             #here employee() is act as function with constructor call

print(f"Content of e1 : {e1.__dict__}")              # "e1" is object, which always contains INSTANCE DATA MEMBER 


e2 = employee(2,"Zoya",999999)                             

print(f"Content of e2 : {e2.__dict__}")             
```

### DIFFERENCES BETWEEN METHODS AND CONSTRUCTORS

===============================

####  METHODS :

===============================


- Methods / Functions are used for perfroming operations on object.
- Methods / Functions names can be any valid variable name.
- Methods / functions must be called Explicitly to execute its definition.
- Methods may or may not return the values.

===============================

#### CONSTRUCTORS :

===============================

- Constructors are used for initializing the objects.
- The name of the constructor is always _ _ init _ _ ( self , ..... ) : pass
- Constructors are automatically / implicitly called by PVM during Object creation.
- Constructor never returns a value ( but it can return None ).
